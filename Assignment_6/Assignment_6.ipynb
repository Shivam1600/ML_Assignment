```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"weather.csv")
df1 = df["MinTemp"]
sns.distplot(df1,hist=True,kde=False)
```

```python
sns.kdeplot(df1,shade=True)
```

```python
sns.distplot(df1,hist=True,kde=True)
```

```python
sns.distplot(df1,hist=False,rug=True)
```

```python
sns.distplot(df1,kde=False,rug=True)
```

```python
sns.distplot(df1,kde=False,hist=False,rug=True)
```

```python
sns.kdeplot(df1,label='original')
sns.kdeplot(df1,bw=3,label='bw=3')
sns.kdeplot(df1,bw=0.7,label='bw=0.7')
plt.legend()
```

```python
df2 = pd.DataFrame(df, columns = ['WindSpeed9am','WindGustSpeed'])
df2.plot(kind='scatter', x='WindGustSpeed', y = 'WindSpeed9am')
```

```python
sns.jointplot(x='WindGustSpeed',y='WindSpeed9am',data=df2)
```

```python
with sns.axes_style('white'):
    sns.jointplot(x = df['Pressure9am'], y =df['Pressure3pm'], kind = 'hex', color = 'k', data=df2)
```

```python
sns.jointplot(x=df['Pressure9am'],y=df['Pressure3pm'],kind='kde')
```

```python
df3 = pd.DataFrame(df, columns=['Humidity3pm','Humidity9am','MinTemp','MaxTemp'])
sns.pairplot(df3)
```
