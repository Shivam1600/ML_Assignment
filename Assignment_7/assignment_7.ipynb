# Univariate Analysis

```python
df[['pH','alcohol']].hist(figsize=(10,4))
```

## Note:-
### pH seems to be almost following normal distribution. 

```python
sns.kdeplot(df['alcohol'])
```

## Note:-
### The value of alcohol increases linearly until it reaches its maximum vaule.
### After alcohol reaches its maximum value i.e., 0.35, its value decreases non linearly.

```python
sns.boxplot( data=df['sulphates'])
```

## Note:-
## 25% value lie in the range of 0.2-0.4 and the minimum value appears to be aroud 0.2
## 50% values lie in between 0.4-0.6.
## More than 75% value lie in the range 0.6-0.8
## There are several outlier values which lie above 0.8

```python
sns.kdeplot(df['pH'], label='original')
sns.kdeplot(df['pH'], bw=0.2, label='bw=0.2')
sns.kdeplot(df['pH'], bw=2, label='bw=2')
plt.legend()
```

## Note:-
### The above graph shows that the highest value of pH is around 3.0

# Bivariate Analysis

```python
df.plot(kind='scatter', x = 'free sulfur dioxide', y = 'total sulfur dioxide')
```

## Note:-
### We can observe that free sulphur dioxide and total sulphur dioxide follows almost linear growth i.e., thwy are almost directly proprtional to one another. 

sns.jointplot(x='fixed acidity', y='alcohol', data=df)

## Note:-
### The density of fixed acidity is most between 6-8 when alcohol values rise from 8-13.

```python
sns.jointplot(x = df['pH'], y =df['density'], kind='hex',  color='R')
```

## Note:-
### Most values of density lies between 0.99-1.00 .
### Most values of pH lies between 3.0-3.4 .
### The histogram plot of pH shows that it almost follows normal distribution.
### The mid value region of both is the most dense.

```python
sns.jointplot(x = df['residual sugar'], y =df['alcohol'], kind='hex',  color='B')
```

## Note:-
### With the increase in residual sugar, alcohol decreases rapidly.
###  The most dense region is when residual sugar is almost 0 and alcohol is around 11

```python
sns.jointplot(x='pH', y='quality', data=df, kind='kde')sns.barplot(x='quality', y='density', data=df)
```

```python
sns.barplot(x='quality', y='chlorides', data=df)
```

## Note:-
### For quality of value 3, chlorides has the maximum number of outlier values
### For quality having values between 5-7, chlorides have minimum number of outlier values

```python
sns.countplot(x='quality', data=df, palette='Greens_d')
```

## Note:-
### White wine has been given the quality of 6 the most and 9 the least number of times.
### White wine can be stated as an average quality product

```python
sns.swarmplot(x ='quality' , y='sulphates', data=df)
```

## Note:-
### Quality having value 7 have more number of sulphates as compared to the other respective values of quality.
### From the previous graph we observed that white wine has had quality of 6 the most and  from this graph we can conclude that more number of sulphates in a samlple results in average quality of around 5-7.
